```python
import pickle
import pandas as pd
```

```python
with open('summary_dict.pkl', 'rb') as fin:
    summary_dict = pickle.load(fin)
```

```python
summary_dict
```

```python
output = ''
TP_all = 0
FP_all = 0
FN_all = 0
```

```python

for k, v in sorted(summary_dict.items()):
    output += 'Note: ' + k + '\n'
    #output += "Script filtered: " + str(filtered_count) + '\n'
    output += "True positive: " + str(v['true_positive']) + '\n'
    output += "False Positive: " + ' '.join(v['false_positive']) + '\n'
    output += "FP number: " + str(len(v['false_positive'])) + '\n'
    output += "False Negative: " + ' '.join(v['false_negative']) + '\n'
    output += "FN number: " + str(len(v['false_negative'])) + '\n'
    if v['true_positive'] == 0 and len(v['false_negative']) == 0:
        output += "Recall: N/A\n"
    else:
        output += "Recall: {:.2%}".format(v['true_positive']/(v['true_positive']+len(v['false_negative']))) + '\n'
    if v['true_positive'] == 0 and len(v['false_positive']) == 0:
        output += "Precision: N/A\n"
    else:
        output += "Precision: {:.2%}".format(v['true_positive']/(v['true_positive']+len(v['false_positive']))) + '\n'
    output += '\n'
    TP_all += v['true_positive']
    FP_all += len(v['false_positive'])
    FN_all += len(v['false_negative'])
    #print(output)
```

```python
#output = "{} notes have been evaulated.\n".format(processed_count-len(miss_file))
output += "True Positive in all notes: " + str(TP_all) + '\n'
output += "False Positive in all notes: " + str(FP_all) + '\n'
output += "False Negative in all notes: " + str(FN_all) + '\n'

if TP_all == 0 and FN_all == 0:
    output += "Recall: N/A\n"
else:
    output += "Recall: {:.2%}".format(TP_all/(TP_all+FN_all)) + '\n'
if TP_all == 0 and FP_all == 0:
    output += "Precision: N/A\n"
else:
    output += "Precision: {:.2%}".format(TP_all/(TP_all+FP_all)) + '\n'
```

```python
print(output)
```

```python
pd_list = []
for k, v in sorted(summary_dict.items()):
    f_name = k.split('_')[0]
    Recall = v['true_positive']/(v['true_positive']+len(v['false_negative']))
    FN_number = len(v['false_negative'])
    FN_detail  = ' '.join(v['false_negative'])
    pd_list.append({'note_id': f_name, '% recall': Recall, '# of FN': FN_number, '# of actual PHI according to HIPAA':'', 'Details': FN_detail })
```

```python
pd_list
```

```python
summary_pd = pd.DataFrame(pd_list)
```

```python
summary_pd.to_csv('FN_summary.csv')
```

```python
with open('category.dict', 'rb') as fin:
    category_dict = pickle.load(fin)
```

```python
pd_list
```

```python
age_list = [] 
date_list = []
id_list = []
location_list = []
name_list = []
contact_list = []
other_list = []
for i in pd_list:
    f_name = i['note_id']+'_phi_reduced'
    fn_list = i['Details'].split(' ')
    for word in fn_list:
        if word != '':
            if word in category_dict[f_name]['age']:
                age_list.append(word)
            elif word in category_dict[f_name]['date']:
                date_list.append(word)
            elif word in category_dict[f_name]['id']:
                id_list.append(word)
            elif word in category_dict[f_name]['location']:
                location_list.append(word)
            elif word in category_dict[f_name]['name']:
                name_list.append(word)
            elif word in category_dict[f_name]['contact']:
                contact_list.append(word)
            else:
                other_list.append([f_name, word])
```
